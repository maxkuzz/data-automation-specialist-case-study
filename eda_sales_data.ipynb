```python
import pandas as pd

orders = pd.read_csv("sales_order.csv")
items = pd.read_csv("sales_order_item.csv")
```

```python
print(orders.head())
print(orders.info())
```

```python
print(items.head())
print(items.info())
```

```python
orders["created_at_dt"] = pd.to_datetime(orders["created_at"], errors="coerce")

print(orders.country_code.value_counts())
print(orders.groupby(["country_code", "currency"]).size())
print(orders.created_at_dt.min(), orders.created_at_dt.max())
```

```python
print(orders.isna().sum())
print("---------")
print(items.isna().sum())
```

```python
zero_or_negative_price = items["product_price_local_currency"] <= 0
zero_or_negative_cost = items["product_cost_eur"] <= 0
non_positive_qty = items["sold_qty"] <= 0
missing_product = items["fk_item"].isna()

print("zero/negative price:", zero_or_negative_price.sum())
print("percent of free items:", round(zero_or_negative_price.mean() * 100, 2), "%\n")

print("zero/negative cost:", zero_or_negative_cost.sum())
print("percent of zero/negative cost:", round(zero_or_negative_cost.mean() * 100, 2), "%\n")

print("qty <= 0:", non_positive_qty.sum())
print("percent of qty <= 0:", round(non_positive_qty.mean() * 100, 2), "%\n")

print("missing product:", missing_product.sum())
print("percent of missing product:", round(missing_product.mean() * 100, 2), "%")
```
